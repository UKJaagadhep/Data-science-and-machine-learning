# **EXPLORATORY DATA ANALYSIS**

**Importing pandas and and reading the required csv file as a dataframe**

```python
import pandas as pd
data=pd.read_csv("/content/Sales_April_2019.csv")
display(data)
```

**Describing the dataframe**

```python
display(data.describe())
```

**Handling missing data**

1) Drop NULL or missing values:

```python
display(data.dropna())
```

2) Fill missing values:

```python

data['Order ID']=data['Order ID'].fillna(0)
data['Product']=data['Product'].fillna(0)
data['Quantity Ordered']=data['Quantity Ordered'].fillna(0)
data['Price Each']=data['Price Each'].fillna(0.0)
data['Order Date']=data['Order Date'].fillna(False)
data['Purchase Address']=data['Purchase Address'].fillna(False)
display(data)
```

**Handling Outliers**

1) Boxplot

```python
import seaborn as sn
import matplotlib.pyplot as pyl
data["Price Each"]=pd.to_numeric(data["Price Each"],errors='coerce')
box_plot=sn.boxplot(data=data, x="Price Each")
pyl.show(box_plot)

```

2) Scatterplot

```python
data["Price Each"]=pd.to_numeric(data["Price Each"],errors='coerce')
data["Quantity Ordered"]=pd.to_numeric(data["Quantity Ordered"], errors='coerce')
scatter_plot=data.plot.scatter(x="Quantity Ordered",y="Price Each")
pyl.show(scatter_plot)
```

3) Z-Score

```python
import numpy as np
import scipy.stats as sp
numpy_array = np.array(data["Price Each"])
numpy_array = numpy_array[~np.isnan(numpy_array)]
z_score=sp.zscore(numpy_array)
zscores=pd.Series(z_score)
display(zscores)



```

```python
#Outliers
data_z=data.dropna()[z_score>=3]
display(data_z)
```

```python
#Without outliers
data=data.dropna()[z_score<3]
display(data)
```

4) IQR

```python
q1=data["Price Each"].quantile(0.25)
q3=data["Price Each"].quantile(0.75)
iqr=q3-q1
iqr_range_1=q1-1.5*iqr
iqr_range_2=q3+1.5*iqr



```

```python
#data outlier
data_outlier=data[(data["Price Each"]<iqr_range_1) | (data["Price Each"]>iqr_range_2)]
display(data_outlier)
```

```python
#dataframe after handling outliers
data=data[(data["Price Each"]>=iqr_range_1) & (data["Price Each"]<=iqr_range_2)]
display(data)
```

**Understanding relationships and new insights through plots**

1) Histogram

```python
hist_plot=sn.histplot(data=data,x='Quantity Ordered')
pyl.show(hist_plot)
```

2) Heat map

```python
data_hmap=pd.concat((data["Price Each"],data["Product"]),axis=1)
data_hmap_1=data_hmap.set_index('Product')
sn.heatmap(data=data_hmap_1.head(50))
```
