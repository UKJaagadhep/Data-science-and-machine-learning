# **Supervised Machine Leaning: Linear Regression with a randomly generated numpy array**

**Import necessary libraries and modules**

```python
import numpy as n
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as pl
```

**Generate random numpy array**

```python
n.random.seed(0)
x=2.5*n.random.rand(200, 1)
y=3*x+n.random.randn(200, 1)
```

**Creating and training a linear regression modelwith the numpy arrays**

```python
model=LinearRegression()
model.fit(x, y)
```

**Plotting graph to show inear regression**

```python
pl.scatter(x, y, label='Original data')
pl.plot(x, model.predict(x), color='red', label='Linear regression')
pl.title('Linear Regression')
pl.xlabel('x')
pl.ylabel('y')
pl.legend()
pl.show()
```
