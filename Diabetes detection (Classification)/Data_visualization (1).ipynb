```python
import numpy as np
count, bin_edge = np.histogram(data['Age'],bins=20, density=True)
print(count)
print(bin_edge)
pdf = count/sum(count)
print(pdf)
cdf=np.cumsum(pdf)
print(cdf)
```

```python
plt.plot(bin_edge[1:],pdf)
```
