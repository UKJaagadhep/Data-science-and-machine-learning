```python
from ultralytics import YOLO

#LOAD MODEL
model = YOLO("yolov8n.pt")

#PERFORM INFERENCE
results = model("//content//dd0d0c744b986774.jpg")

#GET THE FIRST ELEMENT OF RESULTS LISTS WHICH IS THE BOUNDING BOXES WITH IMAGE
print(results)
result = results[0]
```

```python
#SHOW ANNOTATED RESULTS
result.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# PLOT DETECTION RESULTS ON INPUT IMAGE
annotated_image = result.plot()

annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

plt.imshow(annotated_image_rgb)
plt.axis('off')
plt.show()
```

```python
from ultralytics import YOLO

#LOAD MODEL
model = YOLO("yolov8n.pt")

#PERFORM INFERENCE
results = model(["//content//dd0d0c744b986774.jpg", "//content//a5659ba44eaf480d.jpg"])

#GET THE FIRST ELEMENT OF RESULTS LISTS WHICH IS THE BOUNDING BOXES WITH IMAGE
print(results)
result = results[0]
result_ = results[1]
```

```python
import cv2
import matplotlib.pyplot as plt

# PLOT DETECTION RESULTS ON INPUT IMAGE
annotated_image = result.plot()

annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

plt.imshow(annotated_image_rgb)
plt.axis('off')
plt.show()
```
