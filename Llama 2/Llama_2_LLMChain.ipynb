```python
from langchain.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import warnings
warnings.filterwarnings('ignore')
```

```python
#model = "meta-llama/Llama-2-7b-chat-hf" #Needs meta permission
model = "daryl149/llama-2-7b-chat-hf" #No need of any permission
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model = model,
    tokenizer = tokenizer,
    torch_dtype = torch.bfloat16,
    trust_remote_code = True,
    #trust_remote_code = True parameter allows the pipeline to load and execute custom code from the model repository if necessary.
    #It is useful when models have special processing code that is not part of the standard transformers library.
    device_map = "auto",
    #device_map : This parameter allows for automatic assignment of model parts to available devices (such as CPUs or GPUs).
    #Using "auto" will automatically map the model to available hardware, optimizing for performance.
    max_length = 1000,
    do_sample = True,
    top_k = 10,
    num_return_sequences = 1,
    eos_token_id = tokenizer.eos_token_id
    )
```

```python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature' : 0})
```

```python
prompt = "What would be a good name for a AI based startup dealing with NLP tasks on Indian languages?"
```

```python
print(llm(prompt))
```

```python
prompt_template = PromptTemplate(input_variables = ["cuisine"],
                               template = "I want to open a restaurant for {cuisine} food. Suggest a fency name for this")
```

```python
input_prompt = prompt_template.format(cuisine = "Tamil")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm = llm, prompt = prompt_template, verbose = True)
response = chain.run("Tamil")
print(response)
```
