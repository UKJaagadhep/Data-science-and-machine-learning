```python
loader = PyPDFDirectoryLoader('pdfs')
data = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 20)
text_chunks = text_splitter.split_documents(data)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
embeddings.embed_query("Hello! How are you?")
```

```python
len(embeddings.embed_query("Hello! How are you?"))
```

```python
pc = Pinecone()
index_name = "chat"
```

```python
if index_name not in pc.list_indexes().names():
  pc.create_index(index_name,
                  dimension = 768,
                  metric = "cosine",
                  spec = ServerlessSpec(cloud = 'aws', region = 'us-east-1'))
```

```python
pc.list_indexes().names()
```

```python
namespace = "rag_chatbot"

docsearch = PineconeVectorStore.from_documents(
    documents = text_chunks,
    index_name = index_name,
    embedding = embeddings,
    namespace = namespace
)
```

```python
llm = ChatOpenAI(
    openai_api_key = os.environ.get("OPENAI_API_KEY"),
    model_name = "gpt-3.5-turbo",
    temperature = 0.0
)

qa = RetrievalQA.from_chain_type(
    llm = llm,
    chain_type = "stuff",
    retriever = docsearch.as_retriever()
)
```

```python
query = "Tell me about Geographic Named Entity Recognition by Employing Natural Language Processing and an Improved BERT Model."
```

```python
query_with_knowledge = qa.invoke(query)
query_without_knowledge = llm.invoke(query)

print(query_with_knowledge) #RAG
print()
print(query_without_knowledge)
```
