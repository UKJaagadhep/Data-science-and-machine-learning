# **NUMPY**

```python
import numpy as np
list_arr = [[1,2],[3,4]]
numpy_arr = np.array(list_arr) #to create numpy array from a list
print(numpy_arr)
print(numpy_arr.shape) #to get dimensions of numpy array
print(len(numpy_arr.shape)) #to get the number of dimensions of a numpy array
```

```python
print(list(range(8)))
print(np.arange(8)) #does the same as the list command in above line
```

```python
print(np.zeros((2,4)))
print(np.ones((2,4)))
```

```python
print(np.random.randint(8))
print(np.random.randint(8,size=(3,3,2)))
```

```python
print(np.random.random(3))
```

```python
numpy_arr_int8 = np.array(list_arr, dtype = np.int8)
print(numpy_arr_int8, numpy_arr_int8.dtype)
print(numpy_arr.dtype)
```

```python
numpy_arr_float16=np.array(numpy_arr, dtype = np.float16)
print(numpy_arr_float16, numpy_arr_float16.dtype)
```

```python
numpy_arr_uint32=np.array(numpy_arr, dtype = np.uint32)
print(numpy_arr_uint32, numpy_arr_uint32.dtype)
```

```python
numpy_arr_bool=np.array([[0,1],[2,3]], dtype = np.bool)
print(numpy_arr_bool, numpy_arr_bool.dtype)
```

```python
numpy_arr_complex256=np.array(numpy_arr, dtype = np.complex256)
print(numpy_arr_complex256, numpy_arr_complex256.dtype)
```

```python
print(numpy_arr.astype(np.complex128))
print(numpy_arr.astype(np.complex128).dtype)
```
