```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
data = cv2.imread("/content/photocv2.jpeg")
print(data)
print(data.shape)
print(data.dtype)
```

```python
cv2_imshow(data)
cv2.waitKey(2)
cv2.destroyAllWindows()
plt.imshow(data) #shows in BGR format instead of RGB because cv2 stores in BGR format while matplotlib takes in RGB as arguement
```

```python
data_plt = cv2.cvtColor(data, cv2.COLOR_BGR2RGB)
plt.imshow(data_plt)
```

cv2 gives more accurate image than matplotlib

```python
data_gray_bgr = cv2.cvtColor(data, cv2.COLOR_BGR2GRAY)
plt.imshow(data_gray_bgr)
```

```python
cv2_imshow(data_gray_bgr)

```

```python
data_hsv = cv2.cvtColor(data, cv2.COLOR_BGR2HSV)
cv2_imshow(data)
```

```python
b, g, r = cv2.split(data)
b+=30
g+=30
r+=30
data_ = cv2.merge([b,g,r])
cv2_imshow(data_)
```

```python
#Edge detection (ensure image in grayscale)
data_edge = cv2.Sobel(data_gray_bgr, ddepth = cv2.CV_64F, dx = 2, dy = 0) #uses SOBEL filter
cv2_imshow(data_edge)
```
