```python
data=p.read_csv("/content/Summary of Weather.csv", low_memory=False)
display(data)
```

```python
display(data.describe())
```

**Dropping coloumns with only NaN values**

```python
data=data.drop(["FT","FB","FTI","ITH","SD3","RHX","RHN","RVG","WTE","PoorWeather","DR","SPD","WindGustSpd","SND","PGT","TSHDSBRSGF"],axis=1)
display(data)
```

**Filling rest of the NaN values with 0.0**

```python
data.fillna(0.0)
```

**Grouping data and storing them as a new dataframe**

```python
data=data.sort_values('YR')
years=data["YR"].unique()
display(years)
```

```python
data_years=data.groupby(["YR"]).mean(numeric_only=True)
display(data_years)
```

```python
display(data_years.index)
```

```python
data_months=data.groupby(["MO"]).mean(numeric_only=True)
display(data_months)
```

```python
display(data_months.index)
```

**Data Visuaization**

**Mean temperature each day throughout the dataframe 'data'**

```python
scat=s.scatterplot(data=data,x="Date",y="MeanTemp")
pl.pyplot.title("Mean temperature each day by date from year 40 to 45")
display(scat)
```

**Mean temperature each year**

```python
import seaborn as s
import matplotlib as pl
bar_chart_years=s.barplot(data=data_years, x="MeanTemp", y=data_years.index, orient='h')
pl.pyplot.title("Mean Temperature in each year")
pl.pyplot.xlabel('Mean Temperature')
pl.pyplot.ylabel('Year')
display(bar_chart_years)
```

**Mean temperature each month**

```python
bar_chart_months=s.barplot(data=data_months, x=data_months.index, y="MeanTemp")
pl.pyplot.title("Mean Temperature in each month")
pl.pyplot.xlabel('Month')
pl.pyplot.ylabel('Mean Temperature')
display(bar_chart_months)
```

```python
data_years_1=data_years
data_years_1.reset_index(inplace=True)
display(data_years_1)
```

```python
data_months_1=data_months
data_months_1.reset_index(inplace=True)
display(data_months_1)
```

```python
data_grouped=p.melt(data_years,id_vars='YR', value_vars=['MinTemp', 'MaxTemp'],
                      var_name='Category', value_name='Temp',)

display(data_grouped)
```

```python
data_grouped_months=p.melt(data_months_1,id_vars='MO', value_vars=['MinTemp', 'MaxTemp'],
                      var_name='Category', value_name='Temp',)

display(data_grouped_months)
```

**Minimum and maximum temperature each year**

```python
grouped_chart=s.barplot(x='Temp', y='YR', hue='Category', data=data_grouped, palette='Set1',orient="h")
pl.pyplot.xlabel('Temperature')
pl.pyplot.ylabel('Year')
pl.pyplot.title('Minimum and maximum temperatures each year')
pl.pyplot.legend(title='Values')

display(grouped_chart)
```

**Mean maximum and minimum temperatures each month from years 40 to 45**

```python
grouped_chart=s.barplot(x='MO', y='Temp', hue='Category', data=data_grouped_months, palette='Set2')
pl.pyplot.xlabel('Month')
pl.pyplot.ylabel('Temperature')
pl.pyplot.title('Minimum and maximum temperatures each month')
pl.pyplot.legend(title='Values')

display(grouped_chart)
```

**Mean MAX, MIN, MEA by year**

```python
fig,axes=pl.pyplot.subplots(3)
s.lineplot(data=data_years_1, x='YR', y='MIN', ax=axes[0])
s.lineplot(data=data_years_1, x='YR', y='MEA', ax=axes[1])
s.lineplot(data=data_years_1, x='YR', y='MAX', ax=axes[2])
pl.pyplot.xlabel('Year')
```

**Mean MIN, MEA, MAX by month**

```python
data_grouped_months_line=p.melt(data_months_1,id_vars='MO', value_vars=['MIN', 'MEA', 'MAX'],
                      var_name='Category', value_name='Temp',)

display(data_grouped_months_line)
```
